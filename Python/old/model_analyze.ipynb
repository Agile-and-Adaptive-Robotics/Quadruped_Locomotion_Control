```python
import mujoco
import mujoco.viewer
```

```python
dog_model = mujoco.MjModel.from_xml_path('quadruped_model.xml')
dog_data  = mujoco.MjData(dog_model)

actuator_names = [mujoco.mj_id2name(dog_model, mujoco.mjtObj.mjOBJ_ACTUATOR, i) for i in range(dog_model.nu)]

for name in actuator_names:
    print(name)
print("\n")

joint_names = [mujoco.mj_id2name(dog_model, mujoco.mjtObj.mjOBJ_JOINT, i) for i in range(dog_model.njnt)]

for name in joint_names:
    print(name)
print("\n")

geom_names = [mujoco.mj_id2name(dog_model, mujoco.mjtObj.mjOBJ_GEOM, i) for i in range(dog_model.ngeom)]

for name in geom_names:
    print(name)
```
